```python
import matplotlib.pyplot as plt 
#as kısaltma için kullanıldı.
```

```python
x=[1, 2, 3, 4, 5, 6, 7]
y=[5, 10, 18, 23, 28, 34, 39]
a=[1, 2, 3, 4, 5, 6, 7]
b=[4, 15, 19, 21, 25, 39, 40]
```

# LİNE PLOT

```python
plt.plot(x,y)
plt.plot(a,b)
plt.show()
# x,y ve a,b tanımlanması yapıldıktan sonra show tablo halinde birleştirdi.
```

```python
plt.plot(x,y,".")
#Normal de çizgi halinde gösterilirken "." kullanılması dahilinde noktalar birleştirilmeden gösterir.Nokta dışında "+,*" da kullanılabilir.
#"" nokta koymadan denediğimde ise çizgiyi göstermedi.
plt.plot(a,b)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Güneşevler OKULU")
# Tabloya başlık yazmak
plt.ylabel(" SINAV ORTALAMA ")
# y eksenine isim vermek
plt.xlabel(" SINIFLAR ")
# x eksenine isim vermek
plt.show()
```

```python
k=[1,2,3,4]
l=[15,19,12,21]
plt.plot(k,l, label="k sonuç")
plt.plot(x,y, label="x sonuç")
plt.grid()
#çizgi çizmemize yardımcı olur
plt.legend()
#label ile çizgilere isim atadık. legende ile tabloda isimleri gösterdik.
plt.show()
```

# HİSTOGRAM

```python
import seaborn as sns 
# kendi atama yapmadan veri setinde atama yapmamızı sağlar.
```

```python
penguins= sns.load_dataset("penguins")
#Veri setinin tanımlanması yapıldı.
```

```python
penguins.head()
#Veri setinin içine bakmak için kullanılır.
#head() yapıldığında ilk 5 head(sayi) sayi kadar veri gösterir
```

```python
sns.histplot(data=penguins, x=("flipper_length_mm"))
#sns'da x ve ye tanımı yapmadan direk bir veri seti ve ayrıcalıklı veriyi göstererek görselleştirme yapabiliyoruz.
plt.show()
```

# PİE CHART

```python
Kisiler=["ELİF","AYŞE","EDA","CAN","AHMET"]
Satıs=[25,15,40,10,10]
```

```python
plt.pie(Satıs, labels=Kisiler)
#Dilimlerin isimleri labels=isim şeklinde kullanılmalıdır.
#Dilimlere verilen isimler (x,labels=y) y kısmına yazılmalıdır. X kısmına denedim hata veriyor.
plt.show()
```

```python
A=["un","şeker","süt","yumurta"]
B=[10,20,30,40]
plt.pie(B,labels=A, autopct= '%1.f%%')
#autopct: yüzdeleri üstene yazılmasını sağlıyor.
plt.show()
```

# SUBPLOTS

```python
fig= plt.figure()

first_plot= fig.add_subplot(1,3,1)
#add_subplot(Kaç satır grafik, Kaç tane grafik, Kaçıncı grafik )
first_plot.plot(x,y)

second_plot=fig.add_subplot(1,3,2)
second_plot.scatter(x,y, color="red")

third_plot=fig.add_subplot(1,3,3)
third_plot.bar(x,y, color="purple" )


plt.show()
```

```python
fig= plt.figure()

first_plot= fig.add_subplot(3,1,1)
first_plot.plot(x,y)

second_plot=fig.add_subplot(3,1,2)
second_plot.scatter(x,y)

third_plot=fig.add_subplot(3,1,3)
third_plot.bar(x,y)


plt.show()
```

```python
fig= plt.figure(figsize=(18,3))
#figsize grafik boyutunu ayarlamaktadır.

first_plot= fig.add_subplot(1,3,1)
first_plot.plot(x,y,color= "purple")
first_plot.set_title("ÇİZGİ GRAFİĞİ", color="purple")
#Birden fazla grafiğe isim verirken title tek kullanılmaz. set_title olarak kullanılır.

second_plot=fig.add_subplot(1,3,2)
second_plot.scatter(x,y, color="red")
second_plot.set_title("DAĞILIM GRAFİĞİ",color="red")
#Başlığa renk vermek istediğimizde color ile ekleme yapabiliriz.
plt.grid(color="blue")

third_plot=fig.add_subplot(1,3,3)
third_plot.bar(x,y, color="orange" )
third_plot.set_title("ÇUBUK GRAFİĞİ",color="orange")


plt.show()
```
